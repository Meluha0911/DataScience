```python
# define our dataset
dataset = [11,10,12,14,12,15,14,13,15,102,12,14,17,19,107,10,13,12,14,12,108,12,11,14,13,15,10,15,12,10,14,13,15,10]
# this dataset contain an outliers
```

```python
# Using the z_score method to detect an outlier from the dataset
# I will create an empty dictionary to store the outliers
outliers = []
def detect_outliers(data):
    threshold = 3
    mean = np.mean(data)
    std = np.std(data)
    
    for i in data:
        z_score = (i-mean)/std
        if np.abs(z_score)>threshold:
            outliers.append(i)
    return outliers
```

```python
detect_outliers(dataset)
```

```python
sns.boxplot(dataset) # It shows that the data has outliers
```

```python
# IQR  Finding using the code
dataset = sorted(dataset)
dataset # Sort the data 
```

```python
# Calculate Q1 and Q3
q1,q3 = np.percentile(dataset,[25,75])
print(q1,q3)
```

```python
# Find out IQR
IQR = q3-q1
IQR
```

```python
# Find the lower fence and  higher fence
lf = q1-1.5*(IQR)
lf
```

```python
# Find the lower fence and  higher fence
hf = q3+1.5*(IQR)
hf
```

```python
# So my lower fence is 7.5 and the higher fence is 19.5
```

```python
out = []
for i in dataset:
    if i<lf or i>19.5:
        out.append(i)
```
