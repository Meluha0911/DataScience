# How to Plot Bar Plot through Matplotlib

```python
# Basic syntax of the barplot
x = []
y = []
plt.bar(x,y)
plt.show()
```

```python
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y)
plt.show()
```

```python
# How to add label -- yani x aur y ke niche aapko agar koi label lagana hai to wo aap laga sakte ho
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.xlabel("language",fontsize = 15) # Fontsize can also bee change
plt.ylabel("popularity")
plt.title("Sanket") # To give title to the bar graph or whatever graph it may be
plt.bar(x,y)
plt.show()
```

```python
# How to reduce/increase width of the bar
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y,width=0.2)
plt.show()
```

```python
# How to change colour of the bar
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y,color = 'y')
plt.show()
```

```python
# You can also give multiple colour
x = ['python','c','c++','java']
y = [85,70,60,82]
c =['y','b','m','g'] # You can also give the fullname of the colour
plt.bar(x,y,color=c)
plt.show()
```

```python
# How to get python to the edge and not to the center -- bringing the name of x_axis to the edge -- use align parameter
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y,align="edge")
plt.show()
# align ke andar do parameter chalte hai ek hai edge dusra hai center
```

```python
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y,align="center")
plt.show()
# by default align value is center
```

```python
# How to change boundry or edges of the boxes colour
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y,edgecolor = "r")
plt.show()
```

```python
# How to change the size of the edge
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y,edgecolor = 'b',linewidth = 3)
plt.show()
```

```python
# How to change the line style
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y,edgecolor = 'b',linewidth = 3, linestyle = ":")
plt.show()
```

```python
# How to blurr the graph or make it light as it looks more dark -- use the alpha parameter for that purpose
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y,edgecolor = 'b',linewidth = 3,linestyle = ":", alpha = 0.4)
plt.show()
# Alpha ki value 0 se lekar 5 ke bich main hoti hai -- float value.
```

```python
# Label parameter -- to give information about which the graph is -- legend ffunction
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y,edgecolor = 'b',label = 'sanket')
plt.legend()
plt.show()
```

```python
# Other way to give heding to the legend parameter
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y,edgecolor = 'b',linewidth = 3)
plt.legend('PL')
plt.show()
```

```python
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Wscubetech",fontsize=15)
c = ['y','b','m','g']
plt.bar(x,y,color = 'r',edgecolor = 'b',label = 'popularity')
plt.legend()
plt.show()
```

```python
# How to make multiple bar graph -- ekahi graphke andar multiple bar graph kaise banaye
x = ['python','c','c++','java']
y = [85,70,60,82]
z = [20,30,40,50]
plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Wscubetech",fontsize=15)
c = ['y','b','m','g']
plt.bar(x,y,color = 'r',edgecolor = 'b',label = 'popularity')
plt.bar(x,z,color = 'b',label = 'popularity1')
plt.legend()
plt.show()
```

```python
# Yahape humare dono bar graph ek dusarepe overlap hai -- hum unhe overlap karke nahi dikhana hume unhe ek dusre ke pass me
# dikhana hai
# How to make multiple bar graph -- ekahi graphke andar multiple bar graph kaise banaye
x = ['python','c','c++','java']
y = [85,70,60,82]
z = [20,30,40,50]
plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Wscubetech",fontsize=15)
width = 0.2
plt.bar(x,y,width,color = 'r',edgecolor = 'b',label = 'popularity')
plt.bar(x,z,width,color = 'b',label = 'popularity1')
plt.legend()
plt.show()
```

```python
# How to make multiple bar graph -- ekahi graphke andar multiple bar graph kaise banaye
import numpy as np

x = ['python','c','c++','java']
y = [85,70,60,82]
z = [20,30,40,50]

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Wscubetech",fontsize=15)

plt.bar(p,y,color = 'r',label = 'popularity')
plt.bar(p1,z,color = 'y',label = 'popularity1')

plt.legend()
plt.show()
```

```python
import numpy as np

x = ['python','c','c++','java']
y = [85,70,60,82]
z = [20,30,40,50]

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Wscubetech",fontsize=15)

plt.bar(p,y,color = 'r',label = 'popularity')
plt.bar(p1,z,color = 'y',label = 'popularity1')

plt.xticks(p+width/2,x,rotation=10)
plt.legend()
plt.show()
```

```python
# How to make Horizontal bar graph
x = ['python','c','c++','java']
y = [85,70,60,82]
z = [20,30,40,50]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Wscubetech",fontsize=15)

plt.barh(x,y,color = 'r',label = 'popularity')
plt.bar(x,z,color = 'b',label = 'popularity1')
        
plt.legend()
plt.show()

# Just give h parameter at the end of bar like barh
```

```python
# How to make both graphs horizontal
x = ['python','c','c++','java']
y = [85,70,60,82]
z = [20,30,40,50]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Wscubetech",fontsize=15)

plt.barh(x,y,color = 'r',label = 'popularity')
plt.barh(x,z,color = 'b',label = 'popularity1')
        
plt.legend()
plt.show()
```

```python
# How to set the bars -- using the bin parameter
l = [10,20,30,40,50,60]
plt.hist(number,color = 'r',bins = l,edgecolor = 'b') # to 10 aur 20 ke bichke data ko collect karke ek bar bana diya & so
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No.")
plt.show()
```

```python
# How to change range yahape 10 to 60 ye range hai mujhe isase badi wali range chahiye then how
plt.hist(number,'auto',(0,100),edgecolor = 'b') # so (0,100) is use to change the range
# auto to make the bins parameter auto
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No.")
plt.show()
```

```python
l = [10,20,30,40,50,60]
plt.hist(number,color = 'r',edgecolor = 'b') 
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No.")
plt.show()
```

```python
# cumulative parameter -- ye parameter aapka jo bhi frequency bar graph hai usko ye reverse kar deta hai jab aap iski value 
# ko -1 kar dete ho
l = [10,20,30,40,50,60]
plt.hist(number,color = 'b',edgecolor='r',cumulative=-1) # Arrange in the descending order
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No.")
plt.show()
```

```python
# Bottom paarameter -- upar aapki bottom 0 se chal rahi hai lekin aap chahate ho ki wo 0 se na chalkar aapke custome no se
# chale then how
l = [10,20,30,40,50,60]
plt.hist(number,color = 'b',edgecolor='r',bottom=10) # so 10 will occur in the place of the 0
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No.")
plt.show() 
```

```python
# How to change the alignment of the values of the x-axis values
l = [10,20,30,40,50,60]
plt.hist(number,color = 'r',bins=l,edgecolor = 'b',align='mid')
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No.")
plt.show()
```

```python
# Upar wala solid histogram hai -- you can chagne the type of the histogram using the histtype parameter
l = [10,20,30,40,50,60]
plt.hist(number,color = 'r',bins=l,edgecolor = 'b',histtype='step') # To yahape step type main histogram aa gaya hai.
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No.")
plt.show()
```

```python
# How to get stepfilled histogram
l = [10,20,30,40,50,60]
plt.hist(number,color = 'r',bins=l,edgecolor = 'b',histtype='stepfilled')
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No.")
plt.show()
```

```python
# Barstep histogram how to get that
l = [10,20,30,40,50,60]
plt.hist(number,color = 'r',bins=l,edgecolor = 'b',histtype='barstacked')
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No.")
plt.show()
```

```python
# In the simple bar format bar
# Barstep histogram how to get that
l = [10,20,30,40,50,60]
plt.hist(number,color = 'r',bins=l,edgecolor = 'b',histtype='bar')
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No.")
plt.show()
```

```python
# How to change the orientation of the histogram
# Barstep histogram how to get that
l = [10,20,30,40,50,60]
plt.hist(number,color = 'r',bins=l,edgecolor = 'b',orientation="horizontal") # by default it is vertical
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No.")
plt.show()
```

```python
# Yahape aap inki width change kar sakte hai
# Barstep histogram how to get that
l = [10,20,30,40,50,60]
plt.hist(number,color = 'r',bins=l,edgecolor = 'b',rwidth=0.8)
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No.")
plt.show()
# So here the width between the bars has been increased.
```

```python
# How to convert it into logerithm
# Barstep histogram how to get that
l = [10,20,30,40,50,60]
plt.hist(number,color = 'r',bins=l,edgecolor = 'b',log=True)
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No.")
plt.show()
```

```python
# Label parameter to give the label
# Barstep histogram how to get that
l = [10,20,30,40,50,60]
plt.hist(number,color = 'r',bins=l,edgecolor = 'b',label="Pytohn")
plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No.")
plt.legend()
plt.show()
```

```python
# How to divide your graph between two parts 
# Barstep histogram how to get that
l = [10,20,30,40,50,60]
plt.hist(number,color = 'r',bins=l,edgecolor = 'b',label="Pytohn")

plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No.")

plt.axvline(45,color = 'b',label='line') #Kahape xvline dikhani hai, color of that line,lable of that line
plt.legend()
plt.show()
```

```python
# How to use grid pattern
l = [10,20,30,40,50,60]
plt.hist(number,color = 'r',bins=l,edgecolor = 'b',label="Pytohn")

plt.title("Histogram")
plt.xlabel("Python")
plt.ylabel("No.")

plt.axvline(45,color = 'b',label='line') #Kahape xvline dikhani hai, color of that line,lable of that line
plt.legend()
plt.grid()
plt.show()
```

```python
# How to change the size of the dotss of the Scatter Plot
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
size = [100,2,4,5,2,10,7]
plt.scatter(day,no,s = size)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('No',fontsize=15)
plt.show()
# Keep sabki size same
```

```python
# Dots ki transperancy blur karni hai -- with the help of alpha parameter -- alpha parameter ki value 0 se 1 ki bichme rehti
# hai i.e 0.0 to 0.9
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
size = [100,200,400,500,200,100,700]
plt.scatter(day,no,s = size,alpha=0.5)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('No',fontsize=15)
plt.show()
```

```python
# Scatter plot ka marker yahape dot hai yaani o shape main hai yadi yahi marker aapko change karna hai then how -- use
# marker parameter
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
size = [100,200,400,500,200,100,700]
plt.scatter(day,no,s = size,alpha=0.5,marker="*")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('No',fontsize=15)
plt.show()
```

```python
# How to change the edgecolour of the marker
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
size = [100,200,400,500,200,100,700]
plt.scatter(day,no,s = size,alpha=0.5,marker="*",edgecolors="b")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('No',fontsize=15)
plt.show()
```

```python
# How to change the width of edgeline of the marker -- use linewidth parameter
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
size = [100,200,400,500,200,100,700]
plt.scatter(day,no,s = size,alpha=0.5,marker="*",edgecolors="b",linewidth=4)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('No',fontsize=15)
plt.show()
```

```python
# yahape hum color de rahe hai like r,g,w,b but yahape hum color code bhi de sakte hai -- scatter plot ke andar humare pass
# ek aur property hoti hai jisko bolte hai c map property -- c map property ke through bhi aap color change kar sakte hai
# lekin uske andar jo color hota hai wo aapko ek range ke andar dena padta hai -- range in the sense 0 se 100 ke bichme
# aapko ek range milti hai us range ke according aap range de sakte hai -- jaise maine yahapar color use kiya hai inki 
# jagah main no ko use kar sakta hu 0 se lekar 100 ke bichame aur phir main uske according color changr kar sakta hun.
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
color = [10,49,30,29,56,20,30]
size = [100,200,400,500,200,100,700]
plt.scatter(day,no,s = size,c=color,cmap="viridis") # Different diff cmap names are there
plt.colorbar() # To show konse range ke andar konsa color show hoga to see that
plt.title("Scatter Plot",fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('No',fontsize=15)
plt.show()
```

```python
# How to give label in the colorbar
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
color = [10,49,30,29,56,20,30]
size = [100,200,400,500,200,100,700]
plt.scatter(day,no,s = size,c=color,cmap="viridis") # Different diff cmap names are there
t = plt.colorbar() # To show konse range ke andar konsa color show hoga to see that
t.set_label("ColorBar",fontsize = 20)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('No',fontsize=15)
plt.show()
```

# Pie Plot in the Matplotlib

```python
# Basic Syntx of the PiePlot
x = []
plt.pie()
plt.show()
# Useful in the expenditure,profit Analysis
# Imp for the ML
```

```python
x = [10,20,30,40]
y = ['c','c++','java','python'] # Here y will act as a label
plt.pie(x)
plt.show()
```

```python
# Label parameter in the piechart
x = [10,20,30,40]
y = ['c','c++','java','python'] # Here y will act as a label
plt.pie(x,labels=y)
plt.show()
```

```python
# Now i want to show part of the pie chart differently -- seperately -- we can use explode parameter
x = [10,20,30,40]
y = ['c','c++','java','python'] # Here y will act as a label
ex = [0.3,0.0,0.0,0.0] # konse part ko kitna explode karna hai pie se base on the indexing -- 0.0 means normal value 
plt.pie(x,labels=y,explode = ex)
plt.show()
```

```python
# How to explode the java
x = [10,20,30,40]
y = ['c','c++','java','python'] # Here y will act as a label
ex = [0.3,0.0,0.4,0.0] # konse part ko kitna explode karna hai pie se base on the indexing -- 0.0 means normal value 
plt.pie(x,labels=y,explode = ex)
plt.show()
```

```python
# Hoow to change the default colors -- pass the color parameter
x = [10,20,30,40]
y = ['c','c++','java','python'] # Here y will act as a label
c = ['r','b','g','y']
ex = [0.3,0.0,0.0,0.0] # konse part ko kitna explode karna hai pie se base on the indexing -- 0.0 means normal value 
plt.pie(x,labels=y,explode = ex,colors=c)
plt.show()
```

```python
# Ba mujhe percentage bhi show karne hai -- ki graph kitne percentage hai mujhe wobhi show karna hai then how
# use autopct parameter
x = [10,20,30,40]
y = ['c','c++','java','python'] # Here y will act as a label
ex = [0.3,0.0,0.0,0.0] # konse part ko kitna explode karna hai pie se base on the indexing -- 0.0 means normal value 
plt.pie(x,labels=y,explode = ex,autopct="%0.1f%%") # yahape maine 0.1 kiya yani 0 ke bad 1 value aayegi agar main yahape 
# 0.2 karta hun to 0 ke bad 2 values aayegi and f is to give float values
plt.show()
```

```python
# How to bring he shadow in the pieplot / make it more shadowed -- use shadow parameter
x = [10,20,30,40]
y = ['c','c++','java','python']
ex = [0.3,0.0,0.0,0.0]
plt.pie(x,labels=y,explode = ex,autopct="%0.1f%%",shadow=True)  

plt.show()
```

```python
# How to increase the radius of the circle -- how to increase the size of the pie chart -- use rdius parameter for that 
# purpose
x = [10,20,30,40]
y = ['c','c++','java','python']
ex = [0.3,0.0,0.0,0.0]
plt.pie(x,labels=y,explode = ex,autopct="%0.1f%%",shadow=True,radius=1.2)  

plt.show()
```

```python
# Humara jo pie chart ban raha hai yahapar uski starting by default c ke nichese 0% se start ho rahi hai but main us circle
# ko rotate karna chahta hun wo 0% se start na hoke mere customise degree se start ho then how -- use startangle parmter
# Ye jo label hai un label parameter ko hum pie chart se door ya phir pass kar sakte hai -- using the labeldistance
# parameter
x = [10,20,30,40]
y = ['c','c++','java','python']
ex = [0.3,0.0,0.0,0.0]
plt.pie(x,labels=y,explode = ex,autopct="%0.1f%%",shadow=True,startangle=90) # By default it is 0 

plt.show()
# In short startangle se hum circle ko rotate kar sakte hai
```

```python
# Hum label ki fontsize change kar sakte hai --textprops parameter -- iske andar aapk jobhi data de wo as a dictionary pass
# ho -- dictionary ke andar main dunga fontsize -- in key value pair form
x = [10,20,30,40]
y = ['c','c++','java','python']
ex = [0.3,0.0,0.0,0.0]
plt.pie(x,labels=y,explode = ex,autopct="%0.1f%%",shadow=True,textprops={"fontsize":15})  

plt.show()
```

```python
# Upar c,c++,java,python ye counter clockwise chal raha hai mujhe ise clockwise chalana hai then how -- use parameter
# counterclock aur uski value False karni hai -- bydefault it is True
x = [10,20,30,40]
y = ['c','c++','java','python']
ex = [0.3,0.0,0.0,0.0]
plt.pie(x,labels=y,explode = ex,autopct="%0.1f%%",shadow=True,textprops={"fontsize":15},counterclock=False)  

plt.show()
```

```python
# How to bring the changes to the edges of the pie chart -- wedgeprops -- iske andar jobhi property aapko deni hai wo aapko
# dictionary ke format main deni padegi
x = [10,20,30,40]
y = ['c','c++','java','python']
ex = [0.3,0.0,0.0,0.0]
plt.pie(x,labels=y,explode = ex,autopct="%0.1f%%",shadow=True,textprops={"fontsize":15},wedgeprops={"linewidth":4})  # Yaha
# maine line ki width bada di hai 
plt.show()
```

```python
x = [10,20,30,40]
y = ['c','c++','java','python']
ex = [0.3,0.0,0.0,0.0]
plt.pie(x,labels=y,explode = ex,autopct="%0.1f%%",shadow=True,textprops={"fontsize":15},wedgeprops={"linewidth":5,"width":
                                                                                                   2})   
plt.show()
```

```python
# How to change the color of the shadow or the edge -- use edge parameter for that purpose
x = [10,20,30,40]
y = ['c','c++','java','python']
ex = [0.3,0.0,0.0,0.0]
plt.pie(x,labels=y,explode = ex,autopct="%0.1f%%",shadow=True,textprops={"fontsize":15},wedgeprops={"linewidth":4,"edgecolor":
                                                                                                   "m"}) 
plt.show()
```

```python
# How to change the properties of the center -- use the center parameter -- uske andar tuple pass karo
x = [10,20,30,40]
y = ['c','c++','java','python']
ex = [0.3,0.0,0.0,0.0]
plt.pie(x,labels=y,explode = ex,autopct="%0.1f%%",shadow=True,textprops={"fontsize":15},wedgeprops={"linewidth":4,"edgecolor":
                                                                                                   "m"},center = (3,3)) 
plt.show()
```

```python
# Yahapar hum humare label ko rotate kar sakte hai -- using the rotate parameter
x = [10,20,30,40]
y = ['c','c++','java','python']
ex = [0.3,0.0,0.0,0.0]
plt.pie(x,labels=y,explode = ex,autopct="%0.1f%%",shadow=True,textprops={"fontsize":15}
        ,wedgeprops={"linewidth":4},rotatelabels=True) # By default it is True
plt.show()
```
