# How to Plot Bar Plot through Matplotlib

```python
# Basic syntax of the barplot
x = []
y = []
plt.bar(x,y)
plt.show()
```

```python
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y)
plt.show()
```

```python
# How to add label -- yani x aur y ke niche aapko agar koi label lagana hai to wo aap laga sakte ho
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.xlabel("language",fontsize = 15) # Fontsize can also bee change
plt.ylabel("popularity")
plt.title("Sanket") # To give title to the bar graph or whatever graph it may be
plt.bar(x,y)
plt.show()
```

```python
# How to get python to the edge and not to the center -- bringing the name of x_axis to the edge -- use align parameter
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y,align="edge")
plt.show()
# align ke andar do parameter chalte hai ek hai edge dusra hai center
```

```python
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y,align="center")
plt.show()
# by default align value is center
```

```python
# How to change boundry or edges of the boxes colour
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y,edgecolor = "r")
plt.show()
```

```python
# How to change the size of the edge
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y,edgecolor = 'b',linewidth = 3)
plt.show()
```

```python
# How to change the line style
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y,edgecolor = 'b',linewidth = 3, linestyle = ":")
plt.show()
```

```python
# How to blurr the graph or make it light as it looks more dark -- use the alpha parameter for that purpose
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y,edgecolor = 'b',linewidth = 3,linestyle = ":", alpha = 0.4)
plt.show()
# Alpha ki value 0 se lekar 5 ke bich main hoti hai -- float value.
```

```python
# Label parameter -- to give information about which the graph is -- legend ffunction
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y,edgecolor = 'b',label = 'sanket')
plt.legend()
plt.show()
```

```python
# How to make multiple bar graph -- ekahi graphke andar multiple bar graph kaise banaye
import numpy as np

x = ['python','c','c++','java']
y = [85,70,60,82]
z = [20,30,40,50]

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Wscubetech",fontsize=15)

plt.bar(p,y,color = 'r',label = 'popularity')
plt.bar(p1,z,color = 'y',label = 'popularity1')

plt.legend()
plt.show()
```

# Pie Plot in the Matplotlib

```python
# Basic Syntx of the PiePlot
x = []
plt.pie()
plt.show()
# Useful in the expenditure,profit Analysis
# Imp for the ML
```

```python
x = [10,20,30,40]
y = ['c','c++','java','python'] # Here y will act as a label
plt.pie(x)
plt.show()
```

```python
# Label parameter in the piechart
x = [10,20,30,40]
y = ['c','c++','java','python'] # Here y will act as a label
plt.pie(x,labels=y)
plt.show()
```

```python
# Now i want to show part of the pie chart differently -- seperately -- we can use explode parameter
x = [10,20,30,40]
y = ['c','c++','java','python'] # Here y will act as a label
ex = [0.3,0.0,0.0,0.0] # konse part ko kitna explode karna hai pie se base on the indexing -- 0.0 means normal value 
plt.pie(x,labels=y,explode = ex)
plt.show()
```

```python
# How to explode the java
x = [10,20,30,40]
y = ['c','c++','java','python'] # Here y will act as a label
ex = [0.3,0.0,0.4,0.0] # konse part ko kitna explode karna hai pie se base on the indexing -- 0.0 means normal value 
plt.pie(x,labels=y,explode = ex)
plt.show()
```

```python
# Basic syntax of the stem plot
x = []
y = []
plt.stem(x,y)
plt.show()
# Ise stem isliye bolte hai ki yahape jo plotting hai wo dot aur linear plot hai -- dot aur linear plot ka ye matlab hai
# ki aapke pass ek particular data ke upar aapko ek dot milega aur unke sath line ke sath joint hua milega to ise hum bolte
# hai stem plot -- jab aap signal ka use karte hai tab use show karnekeliye aap aise plot ka use karte hai.
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y)
plt.show()
```

```python
# How to change the above solid line into the dotted converted line -- pass linefmt parameter
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,linefmt=":")
plt.show()
```

```python
# How to change the marker i.e dot -- using the markerfmt parameter
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,linefmt=":",markerfmt="r+")
plt.show()
```

```python
# Meri upar bottom line 0 sse start ho rahi hai but main chahta hun ki meri bottom line 1 se ya aur kisi aur no se start ho
# then I can do that also -- use bottom parameter
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,linefmt=":",markerfmt="r+",bottom=1) # By default bottom is 0
plt.show()
```

```python
# Now I had to change the colour of the baseline
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,linefmt=":",markerfmt="r+",bottom=1,basefmt="g")
plt.show()
```

```python
# How to pass the label parameter
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,linefmt=":",markerfmt="r+",bottom=1,basefmt="g",label='python')
plt.legend()
plt.show()
```

```python
# # How to change the colour of the lines -- use use_line_collection parameter
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,linefmt=":",markerfmt="r+",bottom=1,basefmt="g",use_line_collection=False)
plt.show()
```

```python
# How to change the orientation upar vertical orientation hai -- hum ise horizontal orientation bhi kar sakte hai --
# use orientation parameter for that purpose
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,linefmt=":",markerfmt="r+",bottom=1,basefmt="g",orientation="horizontal") # By default it is vertical
plt.show()
```

```python
# dbi parameter in savefig -- matlab ek inch ke andar aapko kitne inch chahiye  wo aap de sakte hai -- improve image quality
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color='r')
plt.savefig('line',dpi=2000)
plt.show()
```

```python
# facecolor parameter -- to change the facecolor
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color='r')
plt.savefig('line',dpi=2000,facecolor="g")
plt.show()
```

```python
# Using the format parameter -- to chaange format mujhe png format main na save karkar mujhe ise pdf format main save karna
# hai
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color='r')
plt.savefig('line2.pdf',dpi=2000,facecolor="g") # you can mention extention while giving the name bydefault it is png
plt.show()
```

```python
# How to use transperant parameter -- agar graph ko transperant dikhana hai to
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color='r')
plt.savefig('line',dpi=2000,facecolor="g",transparent=True) # By default it is false
plt.show()
```

```python
# Box inches parameter
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color='r')
plt.savefig('line',dpi=2000,facecolor="g",transparent=True,bbox_inches='tight') 
```

```python
# Using the annotate
x = [1,2,3,4,5]
y = [3,1,5,2,6]
plt.plot(x,y)
plt.title('WSCubeTech')
plt.xlabel('Days',fontsize = 10)
plt.ylabel('Python',fontsize = 10)

plt.text(2,5,"Java",fontsize=15,style="italic",bbox={'facecolor':'red'}) 
plt.annotate("Python",xy=(2,1),xytext=(4,4),arrowprops={'facecolor':'black','shrink':100})
plt.show()
```
